Period in python : (= full stop = dot)

A period or dot notation in Python is a way to access the data or behavior of a particular object type (tells Python that on what we want to run an operation). 

```python
my_var1 = 'Sahil\'s World'              ## \ before ' or " makes the quotation marks/apostrophe included into the string itself.

var2 = "Sahil's World"
var3 = 'Sahil"s World 3 '
var4 = "bro\" hello.."
var5 = var2 + ' add karto ' + var3      ## concatenation of strings

print(my_var1)
print(var4)
print(var5)
print()

a = 4 
print(type(a))
a = "mama"
print(type(a))
```

```python
print(len(var3))                        # Counts the no. of items in an object, here it's length of the string.
print(len(var5))
print(len(var5[:14]), '\n')             # counted from 0 to 13.

print('sahil','boss')                  ##  (,) gives space in the print function
print(5,'sahil')                        #  combining print of number and string together
print(7,10)
print(7,   10)
print()                                # Print function itself by defaultly prints new line.
#                                        print('\n') = prints 2 blank lines 
#                                        print() = prints 1 blank line


a = 'sahil'
b = 'Hi'
print(a + b)                                        # plus sign operator doesnt add space
print('sahil' + 'Hi')
print(a, b)                                         # comma gives space
print(a, ',', b);   print(a + ', ' + b)             # first doesnt work as we want, second works but complicated in longer statements, thus lets learn,

# Formatted strings (String formatting)
c = '{}, {}! How are u doing?'.format(a.upper(), b)           ##  give Placeholders = {} for all variables, then fill in the placeholders with the var as argument inside format method
print(c)
# F-strings (simplified string formatting)
d = f'{a}, {b}! How are u doing?'                     ## you directly fill the {} with the variable in the F-strings.
print(d)
d = f'{a.upper()}, {b}! How are u doing?'              # Also u can write code inside the placeholders.
print(d)
```

```python
#location = index --> starts with 0
print(var2[0])
print(var2[12])
print(var2[2:11])                     # in maths = [  ,  )  = closed, open


# print(var2[20])                         # index error = means called index is out of range of the string


print(var2[0:11])
print(var2[:11])                      # same as above

print(var2[8:12])                     # last is not included
print(var2[8:])                       # till end (including last) = var[8:str_len] != var2[8:-1]

var2_copy = var2[:]                   # same as var2[0:13] = var2 = copy of the string ; 13=length of the string.
print(var2[8:13])                     # index 13 is invalid , but writing in open bracket works.
```

```python
 #  method   or  functions    -- difference (depth topic)
 # Slicing (depth topic)

print(var3.lower())
print(var3.upper())
print(var3.upper().lower())
print(var3[5:].upper(), '\n')              ## combining concepts

print(var3.count('l'))                     # count method takes string as an argument        
print(var3.count("l"))      
print(var3.count('L'))      
print(var3.upper().count('L'))             ## combining methods also works :)

print()
print(var3.find('o'))                       # index of a certain character
print(var3.find('l'))
print(var3.find('L'))                       # invalid character gives (index = -1)
print(var3.find('y'))
print(var3.find('S'))                       # indices start with 0.
print(var3.find('ah'))                      # finds the index of start of the string.

print('\n')                                 # 2 new lines form, due to, 1 = print itself has a new line bydefault, 2 = of \n

print(var1.replace('World', 'Quanta'))      # Replace method
print(var1)
print(var1.replace('Wold', 'Quanta'))       # But if word is not there, it will not change anything.

print()


# title method = Changes the first letter in each word into upper case, and all others into lower case.
# capitalize method = Upper case the first letter in the sentence and change all others into lower case.
var1 = 'you 3, HELLO'
print(var1.title())
print(var1.capitalize())

# in operator = Boolean output, checks if the string is there or not.
print('World' in var1)
```
